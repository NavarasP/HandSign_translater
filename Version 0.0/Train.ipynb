```python
import cv2
import mediapipe as mp
import numpy as np
from scipy.signal import savgol_filter
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Bidirectional, LSTM, Dense

# Initialize MediaPipe Hands
mp_hands = mp.solutions.hands.Hands(static_image_mode=False, max_num_hands=2, min_detection_confidence=0.7)

# Function to extract hand movement from frames
def extract_hand_movement(frame):
    # Use MediaPipe to detect hand landmarks
    results = mp_hands.process(frame)
    hand_landmarks = results.multi_hand_landmarks
    
    if hand_landmarks:
        # Extract 3D coordinates of hand landmarks
        hand_data = []
        for landmarks in hand_landmarks:
            hand_data.append([landmark.x for landmark in landmarks.landmark] + [landmark.y for landmark in landmarks.landmark] + [landmark.z for landmark in landmarks.landmark])

        hand_data = np.array(hand_data)

        # Apply Savitzky-Golay filter to smooth the hand movement
        # hand_data_smoothed = np.apply_along_axis(apply_savitzky_golay_filter, axis=0, arr=hand_data.T).T

        # Calculate spatial displacement between consecutive frames
        hand_movement = np.diff(hand_data, axis=0)

        return hand_movement

    return None

# Rest of the code remains the same as before

# Capture video from the camera
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    
    # Extract hand movement
    hand_movement = extract_hand_movement(frame)
    print(hand_movement)
    
    # Continue with the rest of your code (e.g., model prediction, etc.)

    # Display the frame
    cv2.imshow("Frame", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
