```python
import os
import json
from pytube import YouTube
from moviepy.video.io.ffmpeg_tools import ffmpeg_extract_subclip
```

```python
output_base_directory = "Dataset/train"
os.makedirs(output_base_directory, exist_ok=True)
```

```python
json_file = "MS-ASL\MSASL_train.json"
with open(json_file, "r") as file:
    data = json.load(file)
```

```python
def download_and_trim_video(video_url, start_time, end_time, clean_text):
    try:
        yt = YouTube(video_url)
        stream = yt.streams.filter(only_video=True, file_extension="mp4").first()
        video_filename = stream.download(output_path="temp")

        output_directory = os.path.join(output_base_directory, clean_text)
        os.makedirs(output_directory, exist_ok=True)

        output_filename = f"{len(os.listdir(output_directory)) + 1}.mp4"
        output_path = os.path.join(output_directory, output_filename)

        ffmpeg_extract_subclip(video_filename, start_time, end_time, targetname=output_path)
        os.remove(video_filename)
        print(f"Downloaded and trimmed {output_path}")
    except Exception as e:
        print(f"Error: {str(e)}")

count=0
for entry in data:
    count+=1
    clean_text = entry["clean_text"]
    video_url = entry["url"]
    start_time = entry["start_time"]  
    end_time = entry["end_time"]      
    download_and_trim_video(video_url, start_time, end_time, clean_text)
    if count==200:
        break
```
